```python
simplest['date_posted'].min(), simplest['date_posted'].max()
```

```python
simplest['year'] = simplest['date_posted'].dt.year
simplest['year'].value_counts().sort_index().plot(kind='bar', figsize=(10, 4), title='Projects per Year')
```

We see a sharp increase in project volume starting around 2007, with steady, high-volume activity from 2009â€“2013. Suggests:

Pre-2007: Sparse data, probably too small to be very useful for training.

2007â€“2012: Good consistent data for model learning.

2013: Could be a good cutoff year for testing â€” it has lots of volume but is still a clean boundary.

Train set: All data before 2013

Test set: Data in 2013

This gives a realistic forecasting scenario. train on long historical trend, test on the most recent complete year.

### Decision Tree on Simplest DF

```python
from sklearn.preprocessing import LabelEncoder

# Assuming your DataFrame is called df and it has columns 'projectid' and 'fully_funded_binary'
le = LabelEncoder()
simplest_1['projectid_encoded'] = le.fit_transform(simplest_1['projectid'])

# Now use this encoded column as your feature
#X = df[['projectid_encoded']]
#y = df['fully_funded_binary']
```

```python
simplest_1.head()
```

```python
cutoff_date = '2013-01-01'

train_df_1 = simplest_1[simplest_1['date_posted'] < cutoff_date]
test_df_1 = simplest_1[simplest_1['date_posted'] >= cutoff_date]

print(f"Train size: {len(train_df_1)}")
print(f"Test size: {len(test_df_1)}")
```

```python
# Use date_posted as numeric feature
X_train_1 = train_df_1[['projectid_encoded']].copy()
X_test_1 = test_df_1[['projectid_encoded']].copy()

# Convert to numeric timestamp
# X_train['date_posted'] = pd.to_datetime(X_train['date_posted']).astype('int64') // 10**9
# X_test['date_posted'] = pd.to_datetime(X_test['date_posted']).astype('int64') // 10**9

# Label
y_train_1 = train_df_1['not_fully_funded']
y_test_1 = test_df_1['not_fully_funded']
```

```python
from sklearn.tree import DecisionTreeClassifier

# Instantiate the model with default settings
clf = DecisionTreeClassifier(random_state=42)

# Fit the model
clf.fit(X_train_1, y_train_1)
```

```python
y_pred_1 = clf.predict(X_test_1)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy
print("Accuracy:", accuracy_score(y_test_1, y_pred_1))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test_1, y_pred_1))

# Classification Report
print("Classification Report:\n", classification_report(y_test_1, y_pred_1))
```

```python
from sklearn.metrics import confusion_matrix

# Get confusion matrix
cm = confusion_matrix(y_test_1, y_pred_1)
tn, fp, fn, tp = cm.ravel()

# Calculate specificity
specificity = tn / (tn + fp)
recall = tp / (tp + fn)
fpr = 1 - specificity

print("Specificity (True Negative Rate):", round(specificity, 3))
print("Recall (True Positive Rate):", round(recall, 3))
print("FPR (1-specificity):", round(fpr, 3))
print('tn:', tn, '\n', 'fp:', fp, '\n', 'fn:', fn, '\n', 'tp:', tp)
```

### Simplest Feature Matrix with Decision Tree Interpretation

TN = 92399: Model predicted all true 0s correctly (fully funded projects)

FP = 0: Model never predicted 1 â€” it predicted all samples as 0 or fully funded

FN = 38930: It missed every actual 1

TP = 0: It missed all the actual 1's (i.e. not fully funded projects)
